```python
df.replace('male', '0', inplace=True)
df.replace('female', '1', inplace=True)
df['diabetic'].replace('No', '0', inplace=True)
df['diabetic'].replace('Yes', '1', inplace=True)
df['smoker'].replace('No', '0', inplace=True)
df['smoker'].replace('Yes', '1', inplace=True)

df['region'].replace('northeast', '0', inplace=True)
df['region'].replace('southeast', '1', inplace=True)
df['region'].replace('southwest', '2', inplace=True)
df['region'].replace('northwest', '3', inplace=True)
df
```

```python
df = df.astype(float)
```

```python
import pandas as pd
import statsmodels.api as sm

X = df[['age', 'gender', 'bmi', 'bloodpressure', 'diabetic', 'children', 'smoker', 'region']]
Y = df['claim']


X = sm.add_constant(X)


model = sm.OLS(Y, X).fit()

print(model.summary())
```
